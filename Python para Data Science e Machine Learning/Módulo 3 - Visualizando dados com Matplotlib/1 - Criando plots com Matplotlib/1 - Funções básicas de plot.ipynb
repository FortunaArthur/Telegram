```python
import numpy as np
x = np.linspace(0, 5, 11)
x
```

```python
y = x**2
y
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
```

```python
plt.plot(x, y, "yellow")
```

```python
plt.plot(x, y, "purple")
plt.title("Nome Titulo")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")
plt.subplot(1, 2, 2)
plt.plot(y, x, "g*-")
```
