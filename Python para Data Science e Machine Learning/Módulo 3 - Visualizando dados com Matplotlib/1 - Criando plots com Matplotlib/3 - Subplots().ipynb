```python
fig, axes = plt.subplots()
axes.plot(x, y, "b*-")
axes.set_title("Titulo do SubPlot")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# o axes é umalista contendo os elemtos, e pra manipular
for axis in axes:
    axis.plot(x, y,"green")
```

```python
## OU 
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y,"green")
axes[1].plot(x, y,"red")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# o axes é umalista contendo os elemtos, e pra manipular
for axis in axes:
    axis.plot(x, y,"orange")
# Essa funbção aki é só pra ajustar na tela, é pra deixa bunitin
plt.tight_layout()
```

```python
# manipular tamanho
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(30, 8))

for axis in axes:
    axis.plot(x, y,"p--")
```
