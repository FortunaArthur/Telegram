```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

```python
fig = plt.figure()
# add eixo (axes)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) # Esquerda, Inferior, Largura, Altura
axes.plot(x, y, "black")
axes.set_xlabel("Label X")
axes.set_ylabel("Label Y")
axes.set_title("Titulo")
```

```python
fig = plt.figure()
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Esquerda, Inferior, Largura, Altura

axes2.plot(y, x, "purple")
axes2.set_xlabel("Label X2")
axes2.set_ylabel("Label Y2")
axes2.set_title("Titulo2")
```
