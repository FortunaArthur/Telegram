```python
import seaborn as sns
vouos = sns.load_dataset("flights")
vouos
```

# 0 - Arruma essa Bagaça

```python
vouos.pivot_table(index="year", columns="month", values="passengers")
```

```python
df_vouos = vouos.pivot_table(index="year", columns="month", values="passengers")
```

# 1 - Heatma

```python
sns.heatmap(df_vouos) # isso daki é o mapinha de calor, só q ta pekeno p krl
```

```python
import matplotlib.pyplot as plt
# lembrar q isso aki retorna uma tupla de figura e eixos
fig, ax = plt.subplots(figsize=(20,8)) # ESSE DAKI É 1 TAMANHO BAUM PRA PODER VER,
sns.heatmap(df_vouos)# pode coloar "ax=ax" como parametro + o pandas n apito nd então segue sem msm
```

```python
fig, ax = plt.subplots(figsize=(20,8))
sns.heatmap(df_vouos, annot=True)# assim vc mostrao valor das caixas
```

```python
fig, ax = plt.subplots(figsize=(20,8))
sns.heatmap(df_vouos, annot=True, fmt=".0f")# assim vc formata
```

```python
# vo da uma aumentada no grafico pq o proximo parametro fica meio ruim da froma q ta agr
fig, ax = plt.subplots(figsize=(30,15))
sns.heatmap(df_vouos, annot=True, fmt=".0f", linewidths=2.5) # assim vc coloca espaçamento entro as caixas pra ver melhor
```

```python
fig, ax = plt.subplots(figsize=(30,15))
sns.heatmap(df_vouos, annot=True, fmt=".0f", linewidths=2.5, cmap="magma") #aki vc muda a paleta de cores
```
