```python
import seaborn as sns
tipos = sns.load_dataset("tips")
tipos
```

# 1 - Lmplot

```python
sns.lmplot(data=tipos, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex")
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex", markers=["x", "v"])
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex", markers=["x", "v"], scatter_kws={"s":200})
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex", markers=["x", "v"], col="time")
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex", markers=["x", "v"], col="time", row="day")
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex", markers=["x", "v"], scatter_kws={"s":100}, aspect=0.3)
```

```python
sns.lmplot(data=tipos, x="total_bill", y="tip", hue="sex", markers=["x", "v"], scatter_kws={"s":100}, aspect=2.3)
```
