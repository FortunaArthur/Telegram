```python
import seaborn as sns
# só par agarantir q os plots funcionem dboa no jupyter
%matplotlib inline
# dataset pego na propria biblioteca do seaborn
tipos = sns.load_dataset("tips")
tipos
```

# 1 - Barplot 

```python
sns.barplot(data=tipos, x="sex", y="total_bill")
```

```python
import numpy as np
#tb da pra mostrar o desvio padrão
sns.barplot(data=tipos, x="sex", y="total_bill", estimator=np.std)
```

# 2 - Countplot

```python
sns.countplot(data=tipos, x="day")
```

```python
sns.countplot(data=tipos, x="smoker")
```

```python
sns.countplot(data=tipos, x="sex")
```

# 3 - Boxplot

```python
# bagulho complicado: basicamente ele segrega por todas as contas de kd dia da semana ai ele ordena do menor pro maior e ai ele marca onde kd grafico se distribui e o risco do meio indica a mediana
```

```python
sns.boxenplot(data=tipos, x="day", y="total_bill")
```

```python
sns.boxenplot(data=tipos, x="total_bill", y="day", orient="h")
```

```python
sns.boxenplot(data=tipos, x="day", y="total_bill", hue="sex")
```

# 4 - Violinplot

```python
sns.violinplot(data=tipos, x="day", y="total_bill")
```

```python
sns.violinplot(data=tipos, x="day", y="total_bill", hue="sex")
```

```python
sns.violinplot(data=tipos, x="day", y="total_bill", hue="sex", split=True) # nesse split ele n espelha ele joga metade p kd 1 pra condençar a info
```

```python
sns.swarmplot(data=tipos, x="day", y="total_bill") # esse aki só coloca so valores e agrupa e mostra onde eles tão concentrados, se tiver d+ ele só empilha
```

```python
sns.swarmplot(data=tipos, x="day", y="total_bill", palette="rainbow")
```
