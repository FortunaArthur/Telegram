```python
import seaborn as sns
# só par agarantir q os plots funcionem dboa no jupyter
%matplotlib inline
```

```python
tipos = sns.load_dataset("tips")
tipos
```

# 1 - Distpltos

```python
sns.distplot(tipos["total_bill"])
```

```python
sns.distplot(tipos["total_bill"], kde=False, bins=49)
```

# 2 - Joinplots

```python
sns.jointplot(x="total_bill", y="tip", data=tipos, kind="reg", color="red")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tipos, kind="hex")
```

# 3 - Pairplot

```python
sns.pairplot(tipos)
```

```python
sns.pairplot(tipos, hue="sex")
```

```python
sns.pairplot(tipos, hue="time")
```

```python
sns.pairplot(tipos, hue="day", palette="coolwarm")
```

# 4 - Rug & KDE plot

```python
sns.rugplot(tipos["total_bill"])
```

```python
sns.kdeplot(tipos["total_bill"])
```

```python
sns.kdeplot(tipos["total_bill"])
sns.rugplot(tipos["total_bill"])
```
