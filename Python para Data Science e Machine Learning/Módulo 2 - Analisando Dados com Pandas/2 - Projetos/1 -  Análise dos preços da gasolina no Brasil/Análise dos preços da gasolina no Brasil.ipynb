### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
# SÃO FUKKING 2 E MEIA DA MANHA KRL, JA TO CORINGANDO NESSA PORRA PQP

dfRio = df[df["ESTADO"] == "RIO DE JANEIRO"]

dfRio ["MES"] = dfRio ["DATA FINAL"] .apply(lambda x: x.month)

mesRio = dfRio.groupby("ANO-MES") [["MES", "PREÇO MÉDIO REVENDA"]].last()
mesRio[mesRio["MES"] == 12]
```

```python
# O DISGRAÇADO FEZ ISSO AKI, EU N SEI + QQ TA KOTECENDO
(mesRio[mesRio["MES"] == 12] / mesRio[mesRio["MES"] == 12].shift(1) - 1) * 100
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
MAXI = df.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"].sort_index()
MINI = df.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"].sort_index()

dfDiferenciado = pd.DataFrame()

dfDiferenciado["MAXIMO"] = MAXI
dfDiferenciado["MINIMO"] = MINI
dfDiferenciado["DIFERENCA"] = MAXI - MINI
dfDiferenciado["DIFERENÇA PERCENTUAL"] = (MAXI - MINI) / MINI * 100

dfDiferenciado
```
